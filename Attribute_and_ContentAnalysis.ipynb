```python
import pandas as pd
import matplotlib.pyplot as plt
import re

# Load the Excel file
datasets_xlsx = pd.read_excel("TenDatasets.xlsx")

first_dataset_index = datasets_xlsx.iloc[0]['index']

# Filtering the DataFrame to find the row corresponding to dataset 52 ("Iris")
#dataset_52_row = datasets_xlsx[datasets_xlsx['index'] == 52]
first_dataset_row = datasets_xlsx[datasets_xlsx['index'] == first_dataset_index]

# Extracting the file URL from the 'dataset_file_url' column
first_dataset_file_url = first_dataset_row['dataset_file_url'].iloc[0]

# Load the uploaded Iris dataset into a DataFrame
#iris_dataset = pd.read_csv("iris.txt", header=None)
iris_dataset = pd.read_csv(first_dataset_file_url, header=None)
# Assigning column names for better understanding
iris_dataset.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']

# after alterations on iris.txt file, to insert some negative values
iris_dataset = pd.read_csv("iris.txt", header=None)
iris_dataset.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']

# Function to return only the rows that contain values less than 0 in numerical columns
def find_rows_with_negative_values(df, numerical_columns):
    # Use the 'any' function along axis=1 to find rows where any of the numerical columns has a value less than 0
    condition = (df[numerical_columns] < 0).any(axis=1)
    return df[condition]

numerical_columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

# Find and display the rows with incorrect values in the newly uploaded dataset
incorrect_rows= find_rows_with_negative_values(iris_dataset, numerical_columns)

# Function to find the file URL corresponding to a given dataset index
def find_dataset_file_url(dataset_index, datasets_df):
    """
    Given a dataset index and a DataFrame containing dataset information,
    this function returns the file URL of the dataset corresponding to the given index.
    
    Parameters:
    - dataset_index (int): The index number of the dataset in the DataFrame.
    - datasets_df (DataFrame): The DataFrame containing information about the datasets.
    
    Returns:
    - str: The file URL of the dataset corresponding to the given index.
    """
    # Filter the DataFrame to find the row corresponding to the given dataset index
    dataset_row = datasets_df[datasets_df['index'] == dataset_index]
    
    # Extract the file URL from the 'dataset_file_url' column
    file_url = dataset_row['dataset_file_url'].iloc[0] if not dataset_row.empty else "Dataset index not found"
    
    return file_url

# Test the function with first dataset index)
find_dataset_file_url(first_dataset_index, datasets_xlsx)

# Analysing third dataset
third_dataset_index = datasets_xlsx.iloc[2]['index']
third_dataset_index

# Load only the 'Age' column (the first column) from the Adult dataset for analysis
# We'll read the file assuming the separator is a comma followed by any number of spaces - this file contains errors in the first two lines
adult_age_data = pd.read_csv("adult.txt", usecols=[0], sep=',\s*', engine='python', header=None)

# Simulate an error by introducing 2 negative values in the 'Age' column
# The first error is at adult_age_data.at[0, 0] = -39  # The 'Age' column is the first (0th) column in adult_age_data

# Function to check if all values in the 'Age' column are numbers between 0 and 130
def check_age_values(column):
    incorrect_values = column[(column < 0) | (column > 130)]
    if not incorrect_values.empty:
        return f"False (Incorrect values found at indices: {incorrect_values.index.tolist()})"
    else:
        return "True"

# Apply the function to the 'Age' column
check_result_age = check_age_values(adult_age_data.iloc[:, 0])
check_result_age
#adult_age_data.head()

# Function to find rows in the 'Age' column where the age is outside the range [0, 130]
def find_rows_with_incorrect_age(df, age_column='Age'):
    # Define a condition for age values that are outside the specified range
    condition = (df[age_column] < 0) | (df[age_column] > 130)
    
    # Filter the DataFrame to get rows that meet the condition
    incorrect_rows = df[condition]
    
    return incorrect_rows

# Apply the function to find and display the rows with incorrect age values
incorrect_age_rows = find_rows_with_incorrect_age(adult_age_data, age_column=0)
incorrect_age_rows
```
